```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib.font_manager as fm
font_name = fm.FontProperties(fname="C:/Windows/Fonts/malgun.ttf").get_name()
plt.rc("font", family=font_name)

import matplotlib as mlp
mlp.rcParams["axes.unicode_minus"] = False

import seaborn as sns
```

```python
KP2020 = pd.read_csv("data/KP2020.csv", encoding="euc-kr")
KP2020
```

```python
time = []
for i in range(len(KP2020)):
    time.append(KP2020["RECV_CPLT_DM"][i].split(".")[0])
time
```

```python
KP2020["RECV_CPLT_DM"] = time
KP2020
```

```python
NPA2020 = pd.read_csv("data/NPA2020.csv", encoding="cp949")
NPA2020
```

```python
NPA2020["RECV_CPLT_DT"] = NPA2020["RECV_CPLT_DT"].astype("str")
NPA2020["RECV_CPLT_TM"] = NPA2020["RECV_CPLT_TM"].astype("str")
NPA2020.info()
NPA2020
```

```python
time = []
for i in range(len(NPA2020)):
    time.append("0"*(6-len(NPA2020["RECV_CPLT_TM"][i]))+NPA2020["RECV_CPLT_TM"][i])
time
```

```python
NPA2020["RECV_CPLT_TM"] = time
NPA2020
```

```python
time = []
for i in range(len(NPA2020)):
    a = NPA2020["RECV_CPLT_TM"][i]
    time.append(a[:2]+":"+a[2:4]+":"+a[4:])
time
```

```python
NPA2020["RECV_CPLT_TM"] = time
NPA2020
```

```python
time = []
for i in range(len(NPA2020)):
    a = NPA2020["RECV_CPLT_DT"][i]
    time.append(a[2:4]+"/"+a[4:6]+"/"+a[6:])
time
```

```python
NPA2020["RECV_CPLT_DT"] = time
NPA2020
```

```python
NPA2020["RECV_CPLT_DM"]=NPA2020["RECV_CPLT_DT"]+" "+NPA2020["RECV_CPLT_TM"]
NPA2020
```

```python
# NPA2020["NPA_CL"].value_counts()
# 13 대전청/ 19 충남청

police = []
for i in range(len(NPA2020)):
    if NPA2020["NPA_CL"][i]==13:
        police.append("대전청")
    else:
        police.append("충남청")
        
police
```

```python
NPA2020["RECV_DEPT_NM"]=police
NPA2020
```

```python
NPA2020 = NPA2020[["RECV_DEPT_NM","RECV_CPLT_DM","NPA_CL","EVT_STAT_CD","EVT_CL_CD","RPTER_SEX","HPPN_OLD_ADDR","HPPN_X","HPPN_Y","SME_EVT_YN"]]
NPA2020.columns = KP2020.columns
NPA2020
```

```python
total2020 = pd.concat([NPA2020, KP2020]).reset_index(drop=True)
total2020
```

```python
total2020["RECV_CPLT_DM"] = pd.to_datetime(total2020["RECV_CPLT_DM"], format="%y/%m/%d %H:%M:%S")
total2020.info()
total2020
```

```python
total2020["요일"] = total2020["RECV_CPLT_DM"].dt.day_name()
total2020
```

```python
KP2021 = pd.read_csv("data/KP2021.csv", encoding="euc-kr")
KP2021
```

```python
time = []
for i in range(len(KP2021)):
    time.append(KP2021["RECV_CPLT_DM"][i].split(".")[0])
time
```

```python
KP2021["RECV_CPLT_DM"] = time
KP2021
```

```python
KP2021["RECV_CPLT_DM"] = pd.to_datetime(KP2021["RECV_CPLT_DM"], format="%y/%m/%d %H:%M:%S")
KP2021.info()
KP2021
```

```python
KP2021["요일"]=KP2021["RECV_CPLT_DM"].dt.day_name()
KP2021
```

```python
total2020.to_csv("data/data/2020년_date변경.csv", index=False)
KP2021.to_csv("data/data/2021년_date변경.csv", index=False)
```
